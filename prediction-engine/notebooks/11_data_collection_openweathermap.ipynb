```python
# download last 6 months data using openweather map API
import requests
import pandas as pd
import json
from datetime import datetime
import pytz
```

```python
# get lat long of mumbai
url = 'http://api.openweathermap.org/geo/1.0/direct?q={}&limit=5&appid={}'.format(city,api_key)
```

```python
response = requests.request('GET', url)
json_response = response.json()
json_response
```

```python
lat = '19.0785451'
long = '72.878176'

# set start date and end date
# Naive datetime
start_dt_naive = datetime(2023, 12, 1, 00, 00)
end_dt_naive = datetime(2024, 12, 1, 00, 00)

# Convert to UTC timezone-aware datetime
start_dt_utc = start_dt_naive.replace(tzinfo=pytz.UTC)
end_dt_utc = end_dt_naive.replace(tzinfo=pytz.UTC)
print('start_dt_utc: ', start_dt_utc)
print('end_dt_utc: ', end_dt_utc)

# Get epoch timestamp
start = int(start_dt_utc.timestamp())
end = int(end_dt_utc.timestamp())

print(start)
print(end)
```

```python
url_history = 'http://api.openweathermap.org/data/2.5/air_pollution/history?lat={}&lon={}&start={}&end={}&appid={}'.format(lat, long, start, end, api_key)
response_history = requests.request('GET', url_history)
json_response_history = response_history.json()
json_response_history
```

```python
json_response_history['list']
```

```python
# Features of the air quality
aqi = []
co = []
no = []
no2 = []
o3 = []
so2 = []
pm2_5 = []
pm10 = []
nh3 = []
datetime = []

for data in json_response_history['list']:
    aqi.append(data['main']['aqi'])
    co.append(data['components']['co'])
    no.append(data['components']['no'])
    no2.append(data['components']['no2'])
    o3.append(data['components']['o3'])
    so2.append(data['components']['so2'])
    pm2_5.append(data['components']['pm2_5'])
    pm10.append(data['components']['pm10'])
    nh3.append(data['components']['nh3'])
    datetime.append(data['dt'])
```

```python
# create dataframe of all the features
aqi_df = pd.DataFrame({'Datetime':datetime,'CO':co,'NO':no,'NO2':no2,'O3':o3,'SO2':so2,
                       'PM2_5':pm2_5,'PM10':pm10,'NH3':nh3,'AQI':aqi})
aqi_df
```

```python
aqi_df.to_csv("../data/raw/aqi_data.csv")
```
